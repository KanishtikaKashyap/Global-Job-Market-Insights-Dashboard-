### Job Type vs. Experience Level Distribution

Let's visualize the distribution of 'Experience Level' across different 'Job type' categories to understand job requirements.

```python
plt.figure(figsize=(12, 7))
sns.countplot(data=df, x='Job type', hue='Experience Level', palette='viridis')
plt.title('Distribution of Experience Level by Job Type')
plt.xlabel('Job Type')
plt.ylabel('Number of Jobs')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Experience Level')
plt.tight_layout()
plt.show()
```
